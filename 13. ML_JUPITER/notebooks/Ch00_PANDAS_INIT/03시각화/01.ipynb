```python
import numpy as np
import matplotlib.pyplot as plt

#   (   )
data_age = np.random.normal(loc=35, scale=10, size=500) #  35,  10

df = pd.read_csv("CC GENERAL.csv")
x = df['CUST_ID']
y= df['PAYMENTS']
plt.figure(figsize=(8, 4))
#  
plt.hist(data_age, bins=20, edgecolor='black', color='skyblue', alpha=0.7) 
plt.title('title')
plt.xlabel('x')
plt.ylabel('y')
plt.axvline(np.mean(data_age), color='red', linestyle='dashed', linewidth=1, label=f'mean: {np.mean(data_age):.2f}') #  
plt.legend()
plt.show()
```
