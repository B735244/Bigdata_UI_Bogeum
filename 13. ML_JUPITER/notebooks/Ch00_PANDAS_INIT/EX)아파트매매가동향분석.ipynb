```python
import pandas as pd
# csv 저장 및 불러오기
# df.to_csv("cafe_index.csv") # 내보내기 -> csv에서 보면 한글이 깨져있음. incoding을 제대로 해주어야 함.
temp = pd.read_csv("(월) 평균매매가격_아파트.csv", encoding="cp949")
# 확인: temp.head()
temp.head(50)
```

```python
train = pd.read_csv("(월) 평균매매가격_아파트.csv", encoding="cp949")
test = pd.read_csv("(월) 평균매매가격_아파트.csv", encoding="cp949")
train
```

```python
from sklearn.model_selection import train_test_split

# 1) 데이터 확인
print(temp.shape)  # 전체 데이터 크기 확인

# 2) train:test = 8:2로 나누기 (랜덤)
train, test = train_test_split(temp, test_size=0.2, random_state=42, shuffle=True)

print("Train shape:", train.shape)
print("Test shape:", test.shape)

# 3) 필요한 경우 인덱스 초기화
train = train.reset_index(drop=True)
test = test.reset_index(drop=True)

# 확인
train.head()
```

```python
import matplotlib.pyplot as plt

# 1행 2열의 서브플롯(그래프 2개)과 전체 크기(가로 8, 세로 4 인치) 설정
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# ------------------------- [왼쪽 그래프: 행 개수 비교] -------------------------
# train과 test의 행(레코드) 수를 리스트로 준비
rows = [train.shape[0], test.shape[0]]

# 막대그래프 그리기: 카테고리 라벨 2개, 값 2개
axes[0].bar(['Train_rows', 'Test_rows'], rows)

# 각 막대 위에 비율(%)을 텍스트로 표시
# enumerate(rows) → i: 막대 인덱스(0/1), v: 막대 높이(해당 데이터 수)
for i, v in enumerate(rows):
    # v / sum(rows) → 전체 행 수 대비 해당 세트의 비중
    # ha='center' 중앙 정렬, va='bottom' 막대 꼭대기 바로 위에 표시
    axes[0].text(i, v, f"{v/sum(rows):.1%}", ha='center', va='bottom')

# 서브플롯 제목
axes[0].set_title("Rows")

# ------------------------- [오른쪽 그래프: 열 개수 비교] -------------------------
# train/test의 열(컬럼) 수를 리스트로 준비
cols = [train.shape[1], test.shape[1]]

# 막대그래프 그리기
axes[1].bar(['Train_cols', 'Test_cols'], cols)

# 각 막대 위에 비율(%) 표시
for i, v in enumerate(cols):
    axes[1].text(i, v, f"{v/sum(cols):.1%}", ha='center', va='bottom')

axes[1].set_title("Columns")

# 레이아웃 간격 자동 조정(레이블/제목이 겹치지 않도록)
plt.tight_layout()

# 그래프 화면에 출력
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# --- Rows ---
rows = [train.shape[0], test.shape[0]]
total_rows = sum(rows) if sum(rows) > 0 else 1  # 0-division 방지
axes[0].bar(['Train', 'Test'], rows)
for i, v in enumerate(rows):
    pct = v / total_rows
    axes[0].text(i, v, f"{v:,} ({pct:.1%})", ha='center', va='bottom')  # 천단위 콤마와 퍼센트 동시 표기
axes[0].set_title("Row counts")

# --- Columns ---
cols = [train.shape[1], test.shape[1]]
total_cols = sum(cols) if sum(cols) > 0 else 1
axes[1].bar(['Train', 'Test'], cols)
for i, v in enumerate(cols):
    pct = v / total_cols
    axes[1].text(i, v, f"{v:,} ({pct:.1%})", ha='center', va='bottom')
axes[1].set_title("Column counts")

plt.tight_layout()
plt.show()
```
